```python
# Coorelation.

# Standardized covariance, also known as correlation, is a statistical measure used to determine the strength and
# direction of the linear relationship between two variables. Correlation values range from -1 to 1, 
# where a correlation coefficient of -1 indicates a perfect negative linear relationship, a correlation
# coefficient of 0 indicates no linear relationship, and a correlation coefficient of 1 indicates a perfect positive 
# linear relationship.

# Quick statistical view of data.
df.describe() 
```

```python
df.corr() #Range from -1 to 1 This is the Standadized Covarience

# To determine whether two variables are related and affect each other, we will make can use of the formular below:

# r = (nΣXY - ΣXΣY) / sqrt[(nΣX^2 - (ΣX)^2)(nΣY^2 - (ΣY)^2)]

# where:

# r = correlation coefficient
# n = number of observations
# ΣXY = sum of the product of each X and Y value
# ΣX = sum of all X values
# ΣY = sum of all Y values
# ΣX^2 = sum of the square of each X value
# ΣY^2 = sum of the square of each Y value


# If the correlation coefficient is close to 1 or -1, it indicates a strong positive or negative linear relationship,
# respectively, between the two variables. If the correlation coefficient is close to 0, it indicates
# little to no linear relationship between the two variables.
```

```python
# Remove outliers

# Calculate the IQR for each column

# IQR stands for Interquartile Range. It is a measure of variability in a dataset, specifically the range between the 
# first quartile (25th percentile) and the third quartile (75th percentile).

# To calculate the IQR, we first calculate the median of the dataset, then divide the data into two halves, where 
# the first half includes all values less than or equal to the median, and the second half includes all values greater 
# than or equal to the median. Then, we calculate the median of each half to obtain the first quartile (Q1) and third 
# quartile (Q3), respectively. The IQR is then calculated as the difference between Q3 and Q1.

# The IQR is often used in statistical analysis to identify and remove outliers from a dataset. 
# It is less sensitive to extreme values than the range or standard deviation, making it a more robust measure of variability.

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
# remove outliers
dfTwo = dfTwo[~((dfTwo < (Q1 - 1.5 * IQR)) | (dfTwo > (Q3 + 1.5 * IQR))).any(axis=1)]
```
